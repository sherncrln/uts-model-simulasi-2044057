```python
import pandas as pd
from sklearn.datasets import load_iris

iris = pd.read_csv('survey lung cancer.csv')
```

```python

iris.info()

iris.head()
```

```python
#menghilangkan kolom yang tidak penting
iris['GENDER'].replace(["F", "M"], [0,1], inplace=True)
```

```python
#memisahkan atribut dan label
X = iris[['GENDER', 'AGE', 'SMOKING', 'CHRONIC DISEASE',  'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']]
y = iris['LUNG_CANCER']

#membagi dataset menjadi dta latih & data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.32, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

#membuat model Desicion Tree

tree_model = DecisionTreeClassifier()

#melatih model dengan menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)

#Evaluasi Model
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

y_pred = tree_model.predict(X_test)

acc_score = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_score)
print('classification_report')
print(classification_report(y_pred,y_test))
```

```python
print(tree_model.predict([[1,30,2,2,2,2,2,2]])[0])
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file = "iris_tree.dot",
    feature_names = ['GENDER', 'AGE', 'SMOKING', 'CHRONIC DISEASE',  'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN'],
    class_names = ['YES', 'NO'],
    rounded = True,
    filled = True)
```

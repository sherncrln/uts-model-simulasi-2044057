```python
import pandas as pd
df = pd.read_csv('survey lung cancer.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#drop kolom yang tidak diperlukan
df['GENDER'].replace(["F", "M"], [0,1], inplace=True)
df['LUNG_CANCER'].replace(["YES", "NO"], [1,0], inplace=True)

#jalankan proses one-hot encoding dengan pd.get_dummies()
df = pd.get_dummies(df)
df
```

```python
#pisahkan atribut dan label
predictions = ['GENDER', 'AGE', 'SMOKING', 'CHRONIC DISEASE',  'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']

X = df[predictions]
y = df['LUNG_CANCER']

#lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split
#bagi data menjadi train dan test untuk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.32, random_state=1)
print(len(y_test)) 
```

```python
from sklearn import linear_model

#latih model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train,y_train)
```

```python
#uji akurasi model
model.score(X_test, y_test)
```

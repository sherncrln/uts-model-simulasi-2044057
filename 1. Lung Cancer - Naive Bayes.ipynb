```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#LOAD DATA
df = pd.read_csv('survey lung cancer.csv')
df.head()
```

```python
df.columns = ['GENDER', 'AGE', 'SMOKING', 'YELLOW FINGERS', 'ANXIETY', 'PEER PRESURE', 'CHRONIC DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING','ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN', 'LUNG CANCER']
```

```python
import matplotlib.pyplot as plt
#plot other feature sets
df.hist(figsize = (20, 20))
plt.show()
```

```python
df['GENDER'].replace(["F", "M"], [0,1], inplace=True)
df['LUNG CANCER'].replace(["YES", "NO"], [1,0], inplace=True)
```

```python
df.head()
```

```python
#feature set
x = df[['GENDER', 'AGE', 'SMOKING', 'CHRONIC DISEASE', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']]
```

```python
#table set
y = df['LUNG CANCER']
```

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.32, random_state=42)

print("Number transactions x_train dataset: ", x_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions x_test dataset: ", x_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
#Train Model

model = GaussianNB()
model.fit(x_train, y_train)
```

```python
#Test Model
y_pred = model.predict(x_test)
print('Accuracy : ', accuracy_score(y_test, y_pred))
```
